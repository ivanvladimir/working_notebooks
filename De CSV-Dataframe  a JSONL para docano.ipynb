```python
df.drop_duplicates(subset=['text'])
df.describe()
```

```python
data=[]

for i,row in df.iterrows():
    label="Yes" if row['label']==1 else "No"
    datum={"data": {
    "text": row['text'],
    "ref_id": i+1,
    "meta_info": {
      "timestamp": "2023-06-01 12:00:00.000000",
      "location": "IIMAS",
      "original_proposed_label": label,
    } 
  },
  "predictions": [{
    "result": [{
      "from_name": "irony_annotation",
      "to_name": "text",
      "type": "choices",
      "readonly": False,
      "hidden": False,
      "value": {
        "choices": [label]
      } 
    }]
  }]}

    data.append(datum)
```

```python
import json

# Serializing json
json_object = json.dumps(data, indent=4)
 
# Writing to sample.json
with open("sample.json", "w") as outfile:
    outfile.write(json_object)
    

        
```
